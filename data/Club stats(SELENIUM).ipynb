```python
url="https://www.uefa.com/uefachampionsleague/statistics/clubs/goals/?sortBy=goals"
```

```python
driver=webdriver.Chrome(PATH)

driver.get(url)

from selenium.webdriver.common.by import By
```

```python
tabla=driver.find_element(By.TAG_NAME, 'pk-table-body')

filas=tabla.find_elements(By.TAG_NAME, 'pk-table-row')


data=[]*50


for f in filas:
    
    elemen=f.find_elements(By.TAG_NAME, 'pk-table-cell')
    
    tmp=[]
    
    for e in elemen:
        
        tmp.append(e.text)
        
    data.append(tmp)

    
data
```

```python
cabeceras=driver.find_element(By.TAG_NAME, 'pk-table-header')

cabeceras=[c.text for c in cabeceras.find_elements(By.TAG_NAME, 'pk-table-header-col')]

driver.quit()

cabeceras
```

```python
import pandas as pd
```
