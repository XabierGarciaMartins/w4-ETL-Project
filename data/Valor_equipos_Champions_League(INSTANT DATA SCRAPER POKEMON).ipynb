```python
import pandas as pd
pd.set_option('display.max_columns', None)

df=pd.read_csv('C:/Users/xabie/proyectototal/Val1.csv')

df.head(20)
```

```python
df=df[['links', 'zentriert', 'zentriert 2', 'rechts', 'rechts 2']]

df.head(20)
```

```python
df2=pd.read_csv('C:/Users/xabie/proyectototal/Val2.csv')

df2.head(20)
```

```python
df2=df2[['links', 'zentriert', 'zentriert 2', 'rechts', 'rechts 2']]

df2.head(20)
```

```python
df2.columns=['Equipo', 'Plantilla', 'Edad_media', 'Valor_equipo', 'Valor_medio_jugadores']

df2.head(20)
```

```python
valor_equipo=df3.copy()
```

```python
valor_equipo
```

```python
valor_equipo = valor_equipo.reset_index(drop=True)
```

```python
valor_equipo
```

```python
t = valor_equipo["Valor_equipo"]
```

```python
lst = []
for i in t:
    if "mil mill" in i :
        lst.append(float(i.split(" m")[0].replace(",","."))*1000)
    else :
        lst.append(float(i.split(" m")[0].replace(",",".")))
        
        
       
    
```

```python
lst
```

```python
valor_equipo["Valor_equipo"] =lst
```

```python
valor_equipo
```

```python
x = valor_equipo["Valor_medio_jugadores"]
```
